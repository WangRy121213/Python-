```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from urllib.error import HTTPError
def getTitle(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        print("Page is wrong")
        return None
    try:
        bs0bj = BeautifulSoup(html.read())
        title = bs0bj.title
    except AttributeError as e:
        print("AttributeError")
        return None
    return title
url = input("please input the website!\n")
title = getTitle(url)
if title == None:
    print("Title could not be found")
else:
    print(title)
```

# 页面去重
   注意使用set集合，该集合最大的特点就是不包含重复的元素

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
page = set()
def getLinks(pageUrl):
    html  = urlopen("http://en.wikipedia.org"+pageUrl)
    bs0bj = BeautifulSoup(html)
    i = 0
    for link in bs0bj.findAll("a",href=re.compile("^(/wiki/)")):
        if i <100:
            i += 1
            if 'href' in link.attrs:
                if link.attrs['href'] not in page:
                    newPage = link.attrs['href']
                    print(newPage)
                    page.add(newPage)
                    getLinks(newPage)
            else:
                break
getLinks("")
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import datetime
import random

page = set()
random.seed(datetime.datetime.now())
allExtLinks = set()
allIntLinks = set()

def getAllExternalLinks(siteUrl):
    html  = urlopen(siteUrl)
    bs0bj = BeautifulSoup(html)
    internalLinks = getInternalLinks(bs0bj,splitAddress(siteUrl)[0])
    externalLinks = getExternalLinks(bs0bj,splitAddress(siteUrl)[0])
    for link in externalLinks:
        if link not in allExtLinks:
            allExtLinks.add(link)
    for link in internalLinks:
        if link not in allIntLinks:
            allIntLinks.add(link)
    print("the length of extLinks is:\n",len(externalLinks))
    print("the length of intLinks is:\n",len(internalLinks))

def getInternalLinks(bs0bj,includeUrl):
    internalLinks = []
    for link in bs0bj.findAll("a",href = re.compile("^(/|.*"+includeUrl+")")):
        if link.attrs['href'] is not None:
            if link.attrs['href'] not in internalLinks:
                internalLinks.append(link.attrs['href'])
    return internalLinks
def getExternalLinks(bs0bj,excludeUrl):
    externalLinks = []
    for link in bs0bj.findAll("a",href = re.compile("^(http|www)((?!"+excludeUrl+").)*$")):
        if link.attrs['href'] is not None:
            if link.attrs['href'] not in externalLinks:
                externalLinks.append(link.attrs['href'])
    return externalLinks
#将网站地址分割开来，且提取首地址
def splitAddress(address):
    addressParts = address.replace("https://","").split("/")
    return addressParts
def getRandomExternalLink(starting):
    html = urlopen(starting)
    bs0bj = BeautifulSoup(html)
    externallinks = getExternalLinks(bs0bj,splitAddress(starting)[0])
    if len(externallinks) == 0:
        internallinks = getInternalLinks(bs0bj,starting)
        if len(internallinks) == 0:
            return "stop"
        else:
            return internallinks[random.randint(0,len(internallinks)-1)]
    else:
        return externallinks[random.randint(0,len(externallinks)-1)]
def followExternalOnly(starting):
    externallink = getRandomExternalLink(starting)
    if externallink!="stop":
        print("随机外链是:\n"+externallink)
        followExternalOnly(externallink)
    else:
        print("ending")
followExternalOnly("http://oreilly.com")
```
