```python
from urllib.request import urlretrieve
from bs4 import BeautifulSoup
import re

pattern = re.compile(r'https://img[0-9]$(.jpg|png|gif)')

weburl = "http://www.douban.com/"
webheaders = {'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36'}

req     = urllib.request.Request(url = weburl,headers = webheaders)
webPage = urllib.request.urlopen(req)

bs0bj = BeautifulSoup(webPage)

for link in bs0bj.findAll("div",{"class":"pic"}):
    #print(link.img.attrs['src'])#bs4中利用这种方法找出全部的attrs['src']
    url1 = link.img.attrs['src']
    print(url1)
    try:
        urlretrieve(url1,destFile(url1))
    except:
        print('Fail')
```

```python
#Python自动登录网站程序
import urllib.request
import urllib
import gzip
import http.cookiejar
from urllib import request,parse

#定义一个方法用于生成请求头信息，处理cookie
def getOpener(head):
    cj  = http.cookiejar.CookieJar()
    pro = urllib.request.HTTPCookieProcessor()
    opener = urllib.request.build_opener(pro)
    header = []
    for key,value in head.items():
        elem = (key,value)
        header.append(elem)
    opener.addheaders = header
    return opener
#定义一个方法来解压返回的信息
def ungzip(data):
    try:
        print("正在解压...")
        data = gzip.decompress(data)
        print("解压完毕...")
    except:
        print("未经压缩，无需解压")
    return data

#封装头信息，伪装成浏览器
header = {
    'Connection':'Keep-Alive',
    'Accept-Language':'zh-CN,zh;q=0.8',
    'Accept':'application/json,text/javascript,*/*;q=0.01',
    'User-Agent':'User-Agent: Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:57.0) Gecko/20100101 Firefox/57.0',
    'Accept-Encoding':'gzip,deflate',
    'X-Requested-With':'XMLHttpRequest',
    'Host':'www.17sucai.com'
}
url    = 'http://www.17sucai.com/auth'
opener = getOpener(header)

id = '1525855'
password = '125689'
postDict = {
    'email' :id,
    'password' : password
}
#编码转换                          #附上需要提交的数据 postDict
                                  #Post的数据必须是bytes或者iterable of bytes，
                                  #因此要进行encode编码
postData = urllib.parse.urlencode(postDict).encode()
op = opener.open(url,postData)
data = op.read()
data = ungzip(data)
print(data)
```
